c- As you see, there are at least 3 categorical features in the dataset (Gender, ChestPain, Thal). Let’s ignore these

```python
feat_cols = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']

X = heart_df[feat_cols]

y = heart_df['AHD']

#print(X.head()) # print the first 5 rows
#print(y.head()) # print the first 5 rows

print(X)
print(y)
```

f- Now, we want to use the categorical features as well! To this end, we have to perform a feature engineering process called
OneHotEncoding for the categorical features. To do this, each categorical feature should be replaced with dummy columns in the feature
table (one column for each possible value of a categorical feature), and then encode it in a binary manner such that only one of the dummy columns can take “1” at a time (and
zero for the rest). For example, “Gender” can take two values “m” and “f”. Thus, we need to replace this feature 
(in the feature table) by 2 columns titled “m” and “f”. Wherever we have a male subject, we can put “1” and ”0” in the columns “m” and “f”. Wherever
we have a female subject, we can put “0” and ”1” in the columns “m” and “f”. (Hint: you
will need 4 columns to encode “ChestPain” and 3 columns to encode “Thal”).

```python
gender = pd.get_dummies(heart_df['Gender'])
chestpain = pd.get_dummies(heart_df['ChestPain'])
thal = pd.get_dummies(heart_df['Thal'])

heart_df = heart_df.drop('Gender',axis=1)
heart_df = heart_df.drop('ChestPain',axis=1)
heart_df = heart_df.drop('Thal',axis=1)

heart_df = heart_df.join(gender)
heart_df = heart_df.join(chestpain)
heart_df = heart_df.join(thal)

heart_df
```

```python
y_gpredict_lr = new_logregr.predict(x_gtest)
y_gpredict_dt = new_decesionTreeCla.predict(x_gtest)

print("\n y_gpredict_lr : ",y_gpredict_lr)
print("\n y_gpredict_dt : ",y_gpredict_dt)
```

```python
new_accuracy_lr = accuracy_score(y_gtest,y_gpredict_lr)
new_accuracy_dt = accuracy_score(y_gtest,y_gpredict_dt)

print("Logistic Regression : ",new_accuracy_lr )
print("Decission Tree Classifier : ",new_accuracy_dt )
print("KNN Accuracy : ",new_accuracy_knn )
```

h- Now, repeat part (e) with the new dataset that you built in part (f), but this time using Cross-Validation. Thus, rather 
than splitting the dataset into testing and training, use 10-fold Cross-Validation (as we learned in Lab4) to evaluate the 
classification methods and report the final prediction accuracy.

```python
from sklearn.cross_validation import cross_val_score

k=3

myknn = KNeighborsClassifier(n_neighbors=k)
mylogreg = LogisticRegression()
myDecisionTree = DecisionTreeClassifier()

knn_accu_list = cross_val_score(myknn,new_X,new_y,cv=10,scoring='accuracy')
lr_accu_list = cross_val_score(mylogreg,new_X,new_y,cv=10,scoring='accuracy')
dt_accu_list = cross_val_score(myDecisionTree,new_X,new_y,cv=10,scoring='accuracy')


print(knn_accu_list)
print(lr_accu_list)
print(dt_accu_list)
```

```python
knn_accuracy = knn_accu_list.mean()
lr_accuracy = lr_accu_list.mean()
dt_accuracy = dt_accu_list.mean()

print("KNN accuracy :",knn_accuracy)
print("Logitic Regression accuracy :",lr_accuracy)
print("Decision Tree Classifier accuracy :",dt_accuracy)
```
