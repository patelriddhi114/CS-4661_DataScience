c- Generate the feature matrix and target vector (target is “balance” in this dataset). Then,
normalize (scale) the features (note: don’t normalize the target vector!). To normalize
the data, you can simply use preprocessing.scale(X) from sklearn.

```python
featu_col = ['Income','Limit','Rating','Cards','Age','Education','Married']

X = credit_df[featu_col]
y = credit_df['Balance']


print(X.head())
print(y.head())

#print(X) # Print whole matrix
#print(y)
```

```python
def featu_scalling(X):
    return X/max(X)


credit_df['Income'] = featu_scalling(credit_df['Income'])
credit_df['Limit'] = featu_scalling(credit_df['Limit'])
credit_df['Rating'] = featu_scalling(credit_df['Rating'])
credit_df['Cards'] = featu_scalling(credit_df['Cards'])
credit_df['Age'] = featu_scalling(credit_df['Age'])
credit_df['Education'] = featu_scalling(credit_df['Education'])

credit_df
```

d- Split the dataset into testing and training sets with the following parameters:
test_size=0.24, random_state=4.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.24,random_state=4)

print("X_train.shape :",X_train.shape)
print("y_train.shape :",y_train.shape)


print("X_test.shape :",X_test.shape)
print("y_test.shape :",y_test.shape)
```

e- Use Linear Regression to train a linear model on the training set. Check the coefficients of
the linear regression model. Which feature is the most important? Which feature is the
least important?

```python
from sklearn.linear_model import LinearRegression

myLinreg = LinearRegression()
myLinreg.fit(X_train,y_train)

```

```python
print("\n Linear Regression Intercept : ",myLinreg.intercept_)
print("\n Linear Refression Coefficient : ",myLinreg.coef_)
```

Which feature is the most important? Which feature is the
least important?

ANS : Linear Regression have most important feature is 10.7488698 it is Cards , 
      and Linear Regression have least important feature is -1404.96322442 it is Age.

f- Predict “balance” for the users in testing set. Then, compare the predicted balance with
the actual balance by calculating and reporting the RMSE (as we saw in lab tutorial 4).

```python
y_prediction = myLinreg.predict(X_test)
print(y_prediction)
```

```python
from sklearn  import metrics

mse = metrics.mean_squared_error(y_test,y_prediction)
rmse = np.sqrt(mse)
print("Calculating RMSE :",rmse)
```

g- Now, use 10-fold Cross-Validation to evaluate the performance of a linear regression in
predicting the balance. Thus, rather than splitting the dataset into testing and training,
use Cross-Validation to evaluate the regression performance. What is the RMSE when you
use cross validation?

```python
from sklearn.cross_validation import cross_val_score

myLinreg = LinearRegression()
rmseList = cross_val_score(myLinreg,X,y,cv=10,scoring='neg_mean_squared_error')

print("RMSE List : ",rmseList)
```

```python
rmseListPosi = -rmseList
rmseList = np.sqrt(rmseListPosi)

print("RMSE positive List : ",rmseList )
```
